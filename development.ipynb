```python
import subprocess, random
%run phugpa.py
def very_close(a,b,eps=1e-6):
  ret = abs(a-b) < eps
  print("[Good]" if ret else "[FAILURE]", " Comparing",a,b,a-b)
  return abs(a-b) < eps
def runProcess(exe):    
    p = subprocess.Popen(exe, stdout=subprocess.PIPE, stderr=subprocess.STDOUT)
    while(True):
      retcode = p.poll() #returns None while subprocess is running
      line = p.stdout.readline()
      yield line
      if(retcode is not None):
        break
```

```python
day, month_count = 1, 300
result = list(runProcess(["perl", "test_moon_equ.pl", "d={}".format(day), "n={}".format(month_count)]))[0]
assert(very_close(float(result), _moon_equ(day, month_count)))
```
